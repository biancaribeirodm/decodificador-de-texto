```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/tabela_de_ações.xlsx" , sheet_name="Principal")
df_principal
```

# Nova secção

```python
df_Total_de_acoes = pd.read_excel("/content/tabela_de_ações.xlsx" , sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel("/content/tabela_de_ações.xlsx" , sheet_name="Ticker")
df_Ticker
```

```python
df_Chatgpt = pd.read_excel("/content/tabela_de_ações.xlsx" , sheet_name="Chatgpt")
df_Chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['var_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal.drop(columns=['Código'])
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['var_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica', 'var_inicial':'valor_inicial'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.drop(columns=['Ticker']).copy()
df_principal
```

```python
df_principal = df_principal.merge(df_Chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'}).copy()
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# Calculando o Maior valor
maior = df_principal['variacao_rs'].max()

# Calculando o Menor valor
menor = df_principal['variacao_rs'].min()

# Calculando a Média
media = df_principal['variacao_rs'].mean()

# Calculando a Média de quem subiu
media_subiu = df_principal.loc[df_principal['Resultado'] == 'Subiu', 'variacao_rs'].mean()

# Calculando a Média de quem desceu
media_desceu = df_principal.loc[df_principal['Resultado'] == 'Desceu', 'variacao_rs'].mean()

print("Maior:", maior)
print("Menor:", menor)
print("Média:", media)
print("Média de quem subiu:", media_subiu)
print("Média de quem desceu:", media_desceu)
```

```python
df_principal_subiu = df_principal.loc[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='variacao_rs', title='Variação reais por Segmento')
fig.update_traces(texttemplate='R$ %{value:,.2f}', textposition='inside')
fig.show()
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='variacao_rs', text='variacao_rs', title='Variação reais por Resultado')
fig.update_traces(texttemplate='R$ %{value:,.2f}', textposition='inside')
fig.show()
```
